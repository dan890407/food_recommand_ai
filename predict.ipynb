```python
import json
with open("comments_set.json","r",encoding="UTF-8") as f:
    data=json.load(f)
```

```python
for store in data:
    print(store)
```

```python
from transformers import BertTokenizer, BertConfig, BertForSequenceClassification
import torch
tokenizer = BertTokenizer.from_pretrained("bert-base-chinese")
config = BertConfig.from_pretrained("./trained_model/config.json") 
model = BertForSequenceClassification.from_pretrained("./trained_model/pytorch_model.bin", config=config)

def predict_sent(sentence):
    inputs = tokenizer(sentence, return_tensors="pt")
    outputs = model(**inputs)
    predicts = outputs[0]
    max_val = torch.max(predicts)
    predict_label = (predicts == max_val).nonzero().numpy()[0][1]
    return str(predict_label)
```

```python
result={}
for store in data:
    pos=0
    neg=0
    for i in store:
        if predict_sent(i)=="1":
            pos+=1
        else:
            neg+=1
    total=pos+neg
    result.update({store:{"possitive":pos/total,"negative":neg/total}})
print(result)
```

```python
with open('sent_data.json', 'w+',encoding="UTF-8") as f:
    json.dump(result, f, ensure_ascii=False, indent=4)
```

```python
from transformers import BartTokenizer, BartForConditionalGeneration ,BartConfig
import logging
import datasets
from torch.utils.data import DataLoader
from tqdm.auto import tqdm, trange
import math
import torch
import transformers
from accelerate import Accelerator
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

config = BartConfig.from_pretrained("./summarize_model/new_epoch_11/config.json")
sum_model = BartForConditionalGeneration.from_pretrained("./summarize_model/new_epoch_11/pytorch_model.bin", config=config).to(device)
tokenizer = BartTokenizer.from_pretrained("facebook/bart-base")


def Summarize_model(model, text,max_len=150):

    input_encodings = tokenizer(
                                    [text],
                                    max_length=1024,
                                    return_tensors="pt",
                                    truncation=True
                                ).to(device)


    output_ids = model.generate(**input_encodings, num_beams=3, max_length=max_len,repetition_penalty=2.5 ,length_penalty=1.0,early_stopping=True,no_repeat_ngram_size= 5, num_return_sequences=1)

    return output_ids
```

```python
import googletrans
def language_translate(source_sentence,target_language = "english"):
  translator = googletrans.Translator()
  # 先將想要的目標語言(target_language)轉為英文，以便等等可以用dest指定轉換到的語言
  language_dest = translator.translate(target_language,dest = 'english').text.lower()
  # 如果是要轉成中文，要變成dest看得懂的格式 
  if language_dest == "chinese":
    language_dest = "chinese (traditional)"
  # 將source_sentence丟入，再指定要轉換到的語言(dest)
  output_sentence = translator.translate(source_sentence,dest = language_dest).text
  return output_sentence
```

```python
with open("comments_cluser.json","r",encoding="UTF-8") as f2:
     data_cluster=json.load(f2)
```

```python
from google_trans_new import google_translator
store_cluster_sum={}
translator = googletrans.Translator()
for store,value in data_cluster.items():
    store_clus={}
    print(store+"\n"+100*"-")
    for clus,reviews in value.items():
        first_sum_set=""
        group_sentence=""
        for i,comment in enumerate(reviews):
            group_sentence+=comment
            if i%5==4 or i==len(reviews)-1:
                sent_set=language_translate(group_sentence,"english")
                first_sum=Summarize_model(sum_model,sent_set,75)
                first_sum=tokenizer.decode(first_sum[0], skip_special_tokens=True)
                first_sum_set+=first_sum
                group_sentence=""
        result_sum=Summarize_model(sum_model,first_sum_set,150)
        result_sum=tokenizer.decode(result_sum[0], skip_special_tokens=True)
        final_sum=language_translate(result_sum,"chinese")
        print(f"{clus}:\n{final_sum}")
        store_clus.update({clus:final_sum})
    store_cluster_sum.update({store:store_clus})
print(store_cluster_sum)
```

```python
with open('clussum_data2.json', 'w+',encoding="UTF-8") as tt:
    json.dump(store_cluster_sum, tt, ensure_ascii=False, indent=4)
```
